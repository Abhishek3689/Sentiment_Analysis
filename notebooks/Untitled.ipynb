```python
import pandas as pd
import numpy as np
```

```python
from gensim.models import Word2Vec
```

```python
tweets=[[''' "Women" and "men" can be important social and political categories.'''],[''' For example, all women—regardless of social class—were prohibited from studying at European and U.S. universities until the late nineteenth century.'''],[''' Until recently, all men as a group were subject to the draft for military duty.''']]
```

```python
tweets
```

```python
import re
```

```python
new_list=[]
for lis1 in tweets:
    for word in lis1:
        tokens=re.sub('[^a-zA-Z]',' ',word)
        tokens=tokens.strip().split()
        new_list.append(tokens)
```

```python
new_list
```

```python
cbow_model=Word2Vec(new_list,vector_size=20,window=2,min_count=1)
```

```python
len(cbow_model.wv.index_to_key)
```

```python
cbow_model.wv.index_to_key
```

```python
cbow_model.wv.get_vector('century')
```
